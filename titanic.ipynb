```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.isna().sum()
```

```python
titanic['age'].fillna(titanic['age'].median(), inplace=True)
```

```python
titanic.drop('deck', axis=1, inplace=True)
```

```python
titanic.dropna(inplace=True)
```

```python
titanic.isna().sum()
```

```python
survivors = titanic[titanic['survived'] == 1]
gender_counts = survivors['sex'].value_counts()
plt.pie(
    gender_counts,
    labels=gender_counts.index,
     autopct='%1.1f%%'
)
plt.title('Survivors by Gender')
plt.legend()

```

```python
town_counts = titanic['embark_town'].value_counts()
print(town_counts)
deceased = titanic[titanic['survived'] == 0]
town_counts = deceased['embark_town'].value_counts()
plt.bar(town_counts.index, town_counts.values)
plt.xlabel('Embark Town')
plt.ylabel('Number of Deaths')
plt.title('Deaths by Embark Town')
```

```python
numeric_cols = titanic.select_dtypes(include='number')
corr = numeric_cols.corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap of Numeric Columns')
```

```python

class_counts = titanic['pclass'].value_counts().sort_index()
print(class_counts)
survivors = titanic[titanic['survived'] == 1]
survived_counts = survivors['pclass'].value_counts().sort_index()
plt.bar(survived_counts.index, survived_counts.values)
plt.xlabel('Passenger Class')
plt.ylabel('Number of Survivors')
plt.title('Number of Survivors by Class')
plt.show()
```

```python

sns.violinplot(y='fare', data=titanic)
plt.title('Fare Distribution for All Passengers')
plt.ylabel('Fare')
```

```python
sns.scatterplot(x='age', y='fare', data=titanic)
plt.title('Age vs. Fare Scatter Plot')
```
